```python
import pandas as pd
df = pd.read_csv('NOVO-B.CO.csv')
df
```

```python
df['Volume Difference'] = df.Volume.diff()
```

```python
df['Daily Change %'] = ((df.Close - df.Open) / df.Open * 100).round(2)
```

```python
year_start = df.at[0, 'Open']
current = 883
((current - year_start)/ year_start) * 100
```

```python
start = df.at[0,'Open']
df['YTD'] = [((i - start) / start) * 100 for i in df.Close]
```

```python
df.round(2)
```

```python
import plotly.express as px
fig = px.line(df,
              x=df.Date,
              hover_data={'YTD': True, 'Volume': True},
              labels={df.YTD.round: 'YTD(%)'},
              y=df.Close)
fig.show()
```

```python
import plotly.express as px

# Calculate YTD returns
start = df.at[0, 'Open']
df['YTD'] = ((df['Close'] - start) / start) * 100

# Round the YTD return to two decimal places
df['YTD'] = df['YTD'].round(2)

# Create the plot
fig = px.line(df,
              x='Date',
              y='Close',
              custom_data=['YTD', 'Volume'],  # Include YTD and Volume in custom hover data
              labels={'YTD': 'YTD Return (%)'},  # Label for YTD return in hover
              title='Stock Close Price with YTD Return')

fig.update_traces(hovertemplate='<b>Date</b>: %{x}<br><b>Close</b>: $%{y:.2f}<br><b>YTD Return</b>: %{customdata[0]:.2f}%<br><b>Volume</b>: %{customdata[1]:,}')
fig.show()
```
