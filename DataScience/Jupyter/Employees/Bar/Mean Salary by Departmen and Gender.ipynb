```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

path = '/Users/cemakay/Data/Employee Sample Data.xlsx'
df = pd.read_excel(path).convert_dtypes()
```

```python
cols = df.columns
for index, col in enumerate(cols):
    print(index,':', col)
```

```python
subset = df.iloc[ : , [5,3,9]]
sal = subset.groupby(['Gender', 'Department']).mean().reset_index()
sal
```

```python
departments = sal['Department'].unique()
y = sal['Annual Salary'].values
female = y[:7]
male = y[7:]


ind = np.arange(len(departments))
width = .35

fig, ax = plt.subplots(figsize=(12,6),layout="constrained")

femalebar = ax.bar(ind - width/2, female, width, color='hotpink', label='Female')
malebar = ax.bar(ind + width/2, male, width, color='g', label='Male')

ax.bar_label(femalebar, label_type='edge', padding=-45, fontsize=12, rotation=90, fmt='%.0f')
ax.bar_label(malebar, label_type='edge', padding=-45, fontsize=12, rotation=90)
ax.set_xticks(ind)
ax.set_xticklabels(departments, fontsize=12, rotation=90, ha='left')
ax.legend(frameon=False, ncols=2)
ax.grid(linestyle='dashed', linewidth=.6)
ax.set_xlim(-.4,4.4)
ax.set_ylim(0, 160000)

plt.show()
```
