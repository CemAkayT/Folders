```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

path = '/Users/cemakay/Data/Employee Sample Data.xlsx'
df = pd.read_excel(path).convert_dtypes()
df.head(2)
```

```python
df.columns
```

```python
cities = df['City'].value_counts()
x = cities.index[::-1]
y = cities.values[::-1]

plt.bar(x,y, label='Cities', color='g')

for key, value in enumerate(y):
    plt.text(key, value, str(value), ha='center', va='top', rotation=-90)

plt.xticks(rotation=90, zorder=2, alpha=.8) #color='r')
plt.grid()
plt.title('Where Are We From?', loc='left')
plt.legend(frameon=False)
plt.show()
```

```python
#Another way to do the same

cities = df['City'].value_counts()
x = cities.index[::-1]
y = cities.values[::-1]

bar = plt.bar(x,y, label='Cities', color='g')

plt.bar_label(bar, label_type='edge', padding=-20, rotation=-90)

plt.xticks(rotation=90, zorder=2, alpha=.8) #color='r')
plt.grid()
plt.title('Where Are We From?', loc='left')
plt.legend(frameon=False)
plt.ylabel('Number of Employees', color='g')
plt.show()
```
