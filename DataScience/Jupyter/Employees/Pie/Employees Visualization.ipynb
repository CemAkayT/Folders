```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('/Users/cemakay/Data/Employee Sample Data.xlsx')
```

```python
df = df.iloc[ : , [1,2,3,5,6,7,9,11,12]]
```

```python
# Replace to HR
df['Department'] = df['Department'].replace('Human Resources', 'HR')
```

```python
df.head()
```

```python
# Ethnicity of employees
ethnicity = dict(df['Ethnicity'].value_counts())
labels = ethnicity.keys()
y = ethnicity.values()

explode = []
for i in range(len(y)):
    if i == len(y) -1:
        explode.append(0.2)
    else:
        explode.append(0.02)

plt.style.use('ggplot')
plt.figure(figsize=(6,12))
plt.pie(y, labels=labels, startangle=180, explode=explode,autopct='%1.1f%%')
plt.title('Employees')
plt.legend(title= 'Etchnicity:', loc='upper right', ncols=2)
plt.show()
```

```python
# Average salary across different departments
mean_sal = (df.groupby(['Department'])
                    .agg(Mean_Salary=('Annual Salary', 'mean'))
                    .reset_index()
                    .round(2)
                    .rename(columns={'Mean_Salary': 'Mean Salary'})
                    .sort_values(by='Mean Salary', ascending=False))

departments = np.array(mean_sal['Department'])
mean_salaries = np.array(mean_sal['Mean Salary'])


# Generate random colors for each department
colors = np.random.rand(len(departments), 3)


fig, ax = plt.subplots()
ax.bar(departments, mean_salaries, label=departments, width=0.5 ,color=colors)


# Add numbers on top of the bars
for i, salary in enumerate(mean_salaries):
    plt.text(i, salary*1, salary, ha='center', va='bottom')
    
    
    
ax.set_ylabel('Average Salary')
ax.legend(title='Departments:', ncols=len(departments), loc='upper right')
fig.set_figwidth(10.8)
ax.set_ylim(0,200000)
plt.xticks(rotation=20, ha='right')
plt.show()
```
