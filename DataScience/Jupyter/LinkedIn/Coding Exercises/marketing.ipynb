### Extract customers with higher level of education, no kids and income above 65000
##### isin() is practical when we have multiple conditions to filter

```python
gruppe = df[~(df.Education.isin(['2n Cycle', 'Basic'])) & (df.Kidhome == 0) & (df.Income > 65000)]
```

### Find the average amount of product purhase by each marital status

```python
df.loc[: , 'MntWines':'MntGoldProds'].groupby(df.Marital_Status).mean().reset_index().round(2)
```

### Find the median income of customers based on education & marital status

```python
df.Income.groupby(by=[df.Education, df.Marital_Status]).agg(['mean', 'median', 'min', 'max']).reset_index()
```
