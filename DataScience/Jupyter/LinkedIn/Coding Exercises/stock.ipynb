```python
import pandas as pd
appl = pd.read_csv('stock_data.csv', usecols=[0,1])
```

```python
appl = appl.rename(columns={appl.columns[0]: 'Date'})
```

```python
appl['Date'] = pd.to_datetime(appl['Date'])
```

```python
appl.set_index('Date', inplace=True)
```

```python
# Group by year and get the last price of each year
appl_yearly = appl.groupby(pd.Grouper(freq='YE')).last()
appl_yearly['Yearly Return'] = appl_yearly.AAPL.pct_change()
appl_yearly
```

```python
#Another method to calucalte
appl['Daily Return'] = appl.pct_change()
```

```python
appl.fillna({'Daily Return': 0}, inplace=True)
```

```python
appl
```

```python
appl['Cum Ret'] = (1+ appl['Daily Return']).cumprod()
appl['Cum Ret'] = appl['Cum Ret'] - 1
```
