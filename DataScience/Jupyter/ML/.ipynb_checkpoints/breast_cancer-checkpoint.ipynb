```python
# imports go here
import pandas as pd
```

```python
pd.set_option('display.max_columns', None) # print all columns. Pandas truncate if too many columns
df = pd.read_csv('/Users/cemakay/Data/breast-cancer-wisconsin-data_data.csv')
df.tail()
```

```python
# 1. Preprocess data  
#df.isnull().sum()
df.isna().sum()
```

```python
# Excluding also last column 'Unnamed: 32' with only NaN values. 
# Drop column before dropping rows with dropna() if one column only has NaN.
X = df.drop(columns=['id', 'diagnosis', 'Unnamed: 32']) # include all columns except 'id' and 'diagnosis'. They dont mean anything for the model
X = X.values # convert from Pandas dataframe to numpy array
X.shape
```

```python
#df.dropna(inplace=True) # handle missing data(remove)
```

```python
# Scale data
# Only use this, if the data is outside -10.0 … 10.0 (as a rule of thumb)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
X
```

```python
y = df.iloc[: , 1] # Target column 'diagnosis at index 1
y = y.map({'M': 1, 'B': 0}) # Encode categorical data to binaries
counts = y.value_counts()
print(counts)
y = y.values # convert from Pandas dataframe to numpy array


# Capital X and lower-case y comes from Linear Algebra. 
# The input is often a 2D array (matrix, named X) while the output is often a 1D array (vector, named y)
```

```python
# 2. Split the data
#After preprocessing, split your dataset into two parts: a training set and a testing set.
#An 80:20 split is a common choice, where 80% of the data is used for training the model, and the remaining 20% is used for evaluating its performance.
#You can use functions like train_test_split from scikit-learn to achieve this

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X[:-2],y[:-2],test_size=0.2, random_state=42)
```

```python
lastY = y[-2:]
lastY
```

```python
lastX = X[-2:]
lastX
```

```python
# 3. Select/Create a  Classification Model
# Choose a machine learning model suitable for classification tasks.
# Common choices include: 
# Logistic Regression: it is used for binary classification
# Decision Trees:
# Random Forests:
# Support Vector Machines (SVM): works well for both linear and nonlinear classification tasks
# Gradient Boosting Models

# Instantiate and fit the model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

# 4. Training the Model
#Fit the chosen model on the training data.
model.fit(X_train, y_train)
```

```python
# 5. Evaluating the model's performance on the test set
#After training, evaluate the model's performance using the testing data.
#Common evaluation metrics for classification tasks include accuracy, precision, recall, F1-score, and ROC-AUC.
from sklearn.metrics import accuracy_score
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
print('Accuracy:', accuracy)
```

```python
# 6. Prediction: 
#Once you're satisfied with the model's performance, you can use it to make predictions on new data.
model.fit(lastX, lastY)
newY = model.predict(lastX)


print('True diagnosis: ', *lastY)
print('Predicted diagnosis: ', *newY)
print('Correct prediction?: ', *lastY == newY)
```
