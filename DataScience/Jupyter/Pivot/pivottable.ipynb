```python
import pandas as pd
import numpy as np

index = np.arange(2,10)
category =['A','B', 'C', 'D','A','B', 'C', 'D' ]


data = pd.DataFrame({'Date': ['2023-01-01', '2023-01-01', '2023-05-01', '2023-04-01', '2023-02-01','2023-02-01','2023-02-01','2023-02-02'],
                     'Category': category,
                     'Value': [12,44,21,33,89,16,72,41]}, index=index).sort_values(by='Date')
print(data)
print('--------------------------------')
dfpivot = pd.pivot_table(data=data, values='Value', index='Date', columns='Category', aggfunc='sum', fill_value=0, margins=True, margins_name='Total')
```

```python
import pandas as pd

data = {
    'Date': ['2024-05-01', '2024-05-01', '2024-05-02', '2024-05-02', '2024-05-03','2024-05-03'],
    'Product': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Quantity': [10, 5, 8, 6, 12, 11],
    'Revenue': [100, 75, 120, 90, 150, 125]
}

data = pd.DataFrame(data)

pivot_table_result = data.pivot_table(index='Date', columns='Product', values=['Quantity', 'Revenue'], aggfunc='sum', margins=True, margins_name='Total')


pivot_table_result
```

```python
category = ['A', 'B', 'C', 'D', 'A', 'B', 'C', 'D']
product = ['iPhone', 'Mac', 'Apple Watch', 'iPhone', 'Mac', 'Apple Watch', 'Macbook Pro', 'Mac']
price = [1200, 2000, 500, 500, 400, 200, 500, 300]

apple = pd.DataFrame({'Category': category,
                      'Product': product,
                      'Price': price})
print(apple)
apple_pivot = apple.pivot_table(index='Category', values='Price', columns='Product', aggfunc='sum', margins=True, fill_value=0)
```
