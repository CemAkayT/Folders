```python
import plotly.express as px
import pandas as pd
df = pd.read_csv('results_2022-2023.csv', index_col='Rk')
df
```

```python

#Remove comma and convert column to numeric
df.Attendance = df.Attendance.str.replace(',', '').astype('Int64')
```

```python
#Remove first and last columns
df = df[df.columns[1:-1]]
df
```

```python
df[['Top Scorer', 'Goals']] = df['Top Team Scorer'].str.split(' - ', expand=True)
df
```

```python
df = df.drop(columns='Top Team Scorer', inplace=True)
```

```python
df.Goals = df.Goals.astype('int64')
df.dtypes
```

```python
import plotly.express as px
fig = px.bar(df,
             x=df.Squad,
             y=df.Pts,
             text_auto=True,
             color=df.Pts,
             labels={'Pts': 'Points', 'Squad': 'Team'},
             height=500)


fig.show()
```

#### Expected Goals 

```python
import plotly.express as px
xg = df.nsmallest(df.shape[0], 'xG')

fig = px.bar(xg,
             x=xg.Squad,
             y=xg.xG,
             text_auto=True,
             title='Expected Goals',
             color=xg.xG)

fig.show()
```

#### Actual Goals Scored

```python
goals = df.nsmallest(len(df),'GF')
fig = px.bar(goals,
         x=goals.Squad,
         y=goals.GF,
         color=goals.GF,
         text_auto=True,
         labels={'GF': 'Goals', 'Squad': 'Team'},
         title='Scored Goals')

fig.show()
```

#### Goals Scored, exptected to score and the difference

```python
xDiff = df.nsmallest(len(df), 'GF')
xDiff
diff = xDiff['GF'] - xDiff['xG']

fig = px.bar(xDiff,
             x='Squad',
             y=['GF', 'xG', diff],
             barmode='group',
             text_auto=True,
             labels={'Squad': 'Team'},
             height=600,
             )


fig.show()
```

```python
small = df.nsmallest(len(df), 'Attendance')

fig = px.bar(small,
             x=small.Squad,
             y=small.Attendance,
             color='Attendance',
             color_continuous_scale='ylgn',
             text='Attendance',
             labels={'Squad': 'Team'},
             text_auto=True)

fig.update_xaxes(tickangle=325)
fig.show()
```

```python
#Split 'Team Top Scorer' to 2 columns and drop later the intermediary and not usefull columns
df[['Topscorer', 'Goals']] = df['Top Team Scorer'].str.split(' - ', expand=True)
```

```python
import numpy as np
df['Difference'] = df['GF'] - df['xG']
diff = df.nsmallest(len(df), 'Difference') 
fig = px.bar(diff,
             x=diff.Squad,
             y=diff.Difference,
             text_auto=True,
             text='Difference',
             color=diff.Difference,
             color_continuous_scale='RdYlGn',
             title='Goal Difference between scored and expected')

fig.show()
```

```python
gd = df.nsmallest(len(df), 'GD')
fig = px.bar(gd,
              x=gd.Squad,
              y=gd.GD,
              color=gd.GD,
              text_auto=True,
              title='Goal Difference',
              color_continuous_scale='RdYlGn')

fig.show()
```

```python
gf = df.nsmallest(len(df), 'GF')

fig = px.bar(gf,
             x= gf.GF,
             y=gf.Squad,
             text_auto=True,
             color=gf.GF,
             labels={'Squad': 'Team', 'GF': 'Goals Scored'},
             color_continuous_scale='greens')

# Update layout
fig.update_layout(
    title='Scored Goals 2022/2023',
    height=600,
    width=800,
    xaxis_title='Number of Goals',
    yaxis_title='Team',
    font=dict(family='Arial', size=11, color='black'),
    margin=dict(l=130, r=50, t=80, b=60),
    plot_bgcolor='lightgrey')
fig.show()
```
