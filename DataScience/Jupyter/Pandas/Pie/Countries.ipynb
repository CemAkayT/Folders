```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

path = '/Users/cemakay/Data/Employee Sample Data.xlsx'
df = pd.read_excel(path).convert_dtypes()
```

```python
conts = df['Country'].value_counts()

x = conts.index
y = conts.values

fig, ax = plt.subplots()

ax.pie(y,labels=x, autopct='%1.1f%%', explode=[.01,.01,.2],colors=['blue', 'red', 'green'], textprops={'color':"black", 'fontweight': 'light'})
ax.set_title('Country of Employees', fontsize=12, fontweight='bold', y=1, loc='center')
ax.legend(loc='upper right' , bbox_to_anchor=(1.55, 1.1), ncols=1 )
plt.show()
```
