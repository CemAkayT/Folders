```python
import pandas as pd
```

```python
path = '/Users/cemakay/Data/Employee Sample Data.xlsx'
df = pd.read_excel(path).convert_dtypes()
df.info()
```

```python
# Exclude first and last columns
df = df.iloc[ : , 1:-1]
df.info()
```

```python
#Change name to HR
df['Department'] = df['Department'].replace({'Human Resources': 'HR', 'IT': 'Information Tech'})
```

```python
deps = df['Department'].value_counts()
deps
```

```python
#!pip install PyQt6
#%matplotlib qt
import matplotlib.pyplot as plt
import numpy as np

deps = df['Department'].value_counts()
x = deps.index
y = deps.values

#random colors
colors = np.random.rand(len(deps),3)

explode = []
for i in range(len(y)):
    if i == 0:
        explode.append(.2)
    else:
        explode.append(.01)

fig, ax = plt.subplots(figsize=(6,6))
#plt.style.use('ggplot')
#fig.set_figwidth(10.8)
ax.pie(y, labels=x, startangle=140, colors=colors, explode=explode, autopct='%1.1f%%')
ax.set_title('Employees by Departments', fontsize=12, fontweight='bold', y=1, loc='center')
ax.legend(title= 'Departments:', loc='upper right', bbox_to_anchor=(1.55, 1.1), ncols=1)
plt.show()                     
```
