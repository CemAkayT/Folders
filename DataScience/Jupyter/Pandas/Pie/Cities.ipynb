```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

path = '/Users/cemakay/Data/Employee Sample Data.xlsx'
df = pd.read_excel(path).convert_dtypes()
```

```python
city = df['City'].value_counts()
```

```python
x = city.index
y = city.values

values = np.full(13,.01)
values[0] = .4

plt.subplots(figsize=(12,7))
plt.pie(y, labels=x, autopct='%1.1f%%', explode=values, startangle=180)
plt.title('Where are we from?', fontsize=12, fontweight='bold', y=1, loc='center')
#plt.legend(title='Gender :', loc='upper right', bbox_to_anchor=(1.55, 1.1), ncols=1 )
plt.show()
```
