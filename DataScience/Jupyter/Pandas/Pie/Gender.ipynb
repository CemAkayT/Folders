```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

path = '/Users/cemakay/Data/Employee Sample Data.xlsx'
df = pd.read_excel(path).convert_dtypes()
```

```python
genders = df['Gender'].value_counts()

x = genders.index
y = genders.values

colors = {'Female': 'pink', 'Male': 'red'}

fig, ax = plt.subplots()
ax.pie(y, labels=x, autopct='%1.1f%%', colors=[colors[gender] for gender in x], explode=[.02, .01], startangle=90)
ax.set_title('Gender Distribution', fontsize=12, fontweight='bold', y=1, loc='center')
ax.legend(title='Gender :', loc='upper right', bbox_to_anchor=(1.55, 1.1), ncols=1 )
plt.show()
```
