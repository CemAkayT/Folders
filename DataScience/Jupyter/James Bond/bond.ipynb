```python
import pandas as pd
import numpy as np

df = pd.read_csv('/Users/cemakay/DataScience/Jupyter/Data/james_bond_data.csv')
df.head()
```

```python
df.Film_Length = df.Film_Length.str.replace(' mins', '').astype(int)
```

```python
df.head()
```

```python
df.Bond = df.Bond.str.replace('Shawn', 'Sean').str.replace('MOORE', 'Moore')
```

```python
df.Bond.value_counts()
```

```python
df.head()
```

```python
df['Budget ($ 000s)'] = df['Budget ($ 000s)'].str.replace('[$,.]', '', regex=True).astype(float) * 10
```

```python
df.head()
```

```python
df.US_Gross  = df.US_Gross.str.replace('[$,]','', regex=True).astype(float)
```

```python
df.World_Gross  = df.World_Gross.str.replace('[$,]','', regex=True).astype(float)
```

```python
df.head()
```

```python
df.Bond_Car_MFG = df.Bond_Car_MFG.str.replace('Astin', 'Aston')
```
