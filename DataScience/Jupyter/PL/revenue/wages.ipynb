```python
#!/Users/cemakay/Jupyter/Pandas/.venv/bin/python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#%matplotlib qt


clubs = ['Liverpool', 'Arsenal', 'Man City', 'Aston Villa', 'Tottenham', 'Man United', 
          'West Ham', 'Newcastle', 'Brighton', 'Chelsea', 'Wolves', 'Bournemouth', 'Fulham', 
          'Crystal Palace', 'Brentford', 'Everton', 'N.Forest', 'Luton', 'Burnley', 'Sheffield United']


np.random.seed(1)
values = np.random.randint(50,101,(20,5))
matchday = values[: , 0]
broadcast = values[: , 1]
commercial = values[: , 2]
wages = values[: , 3]
total = (matchday + broadcast + commercial) - wages


df = pd.DataFrame({'Matchday': matchday,
                   'Broadcast': broadcast,
                   'Commercial': commercial,
                   'Wages': wages,
                   'Total': total,
                   },index=clubs).convert_dtypes()

sorted_wages = df.sort_values(by='Wages', ascending=True)



fig, ax  = plt.subplots()
t = ax.bar(sorted_wages.index, sorted_wages['Wages'],color='g', label='Wages')
ax.bar_label(t, label_type='edge', padding=-12, color='w', fontweight='bold', fontsize=8)
ax.set_xticks(sorted_wages.index)
ax.set_xticklabels(sorted_wages.index, rotation=90)
ax.grid()
ax.set_xlim(-0.5 , 19.5)
plt.legend(loc='upper left', frameon=False, labelspacing=.1, ncols=3)
plt.title('\nMatchday, Commercial and Broadcast revenue 2023\n Millions\n', loc='left', fontweight='bold', fontsize=16)
plt.subplots_adjust(top=0.879, bottom=0.212, left=0.051, right=0.99, hspace=0.2, wspace=0.2)
plt.show()
```
