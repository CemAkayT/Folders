```python
#!/Users/cemakay/Jupyter/Pandas/.venv/bin/python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#%matplotlib qt

np.random.seed(2)
clubs = ['Liverpool', 'Arsenal', 'Man City', 'Aston Villa', 'Tottenham', 'Man United', 
          'West Ham', 'Newcastle', 'Brighton', 'Chelsea', 'Wolves', 'Bournemouth', 'Fulham', 
          'Crystal Palace', 'Brentford', 'Everton', 'N.Forest', 'Luton', 'Burnley', 'Sheffield United']

matchday = np.random.randint(500, 1001, 20)
broadcast = np.random.randint(500, 1001, 20)
commercial = np.random.randint(500,1001, 20)


df = pd.DataFrame({'Matchday': matchday,
                     'Broadcast': broadcast,
                     'Commercial': commercial,}, index=clubs).convert_dtypes()


df['Revenue 2023'] = (matchday + broadcast + commercial)
df = df.sort_values(by=['Revenue 2023'], ascending=True)

print(df)

x = np.arange(len(df.index))
match = df['Matchday'].values
broad = df['Broadcast'].values
comm = df['Commercial'].values
width = .3

fig, ax = plt.subplots(figsize=(12,6))


bar1 = plt.bar(x-.3, match, width, color='r', label='Matchday', zorder=2 ,alpha=.8)
bar2 = plt.bar(x, broad, width, color='g', label='Broadcast',zorder=2, alpha=.8)
bar3 = plt.bar(x+.3, comm, width, color='b',label='Commercial', zorder=2, alpha=.8)


plt.xticks(x, df.index, rotation=90)
plt.ylim(0,1100)

plt.subplots_adjust(bottom=0.15)

for bars in [bar1, bar2, bar3]:
    plt.bar_label(bars,color='w', fontweight='bold', fontsize=8, padding=-23, rotation=90)

plt.legend(loc='upper left')

plt.title('Revenue\n Millions', loc='left', fontweight='bold')
plt.grid()
plt.show()
```
