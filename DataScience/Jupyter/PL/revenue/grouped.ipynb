```python
#!/Users/cemakay/Jupyter/Pandas/.venv/bin/python


import numpy as np
import pandas as pd

np.random.seed(2)
clubs = ['Liverpool', 'Arsenal', 'Man City', 'Aston Villa', 'Tottenham', 'Man United', 
          'West Ham', 'Newcastle', 'Brighton', 'Chelsea', 'Wolves', 'Bournemouth', 'Fulham', 
          'Crystal Palace', 'Brentford', 'Everton', 'N.Forest', 'Luton', 'Burnley', 'Sheffield United']
revenue1 = np.random.randint(500,1501, 20)
revenue2 = np.random.randint(500,1501, 20)

df = pd.DataFrame({'Revenue 2022': revenue1,
                   'Revenue 2023': revenue2,
                   'Total': revenue1 + revenue2
                   }, index=clubs)

df = df.sort_values(by=['Total'], ascending=True)
print(df)



import matplotlib.pyplot as plt
x = np.arange(len(df.index))
rev_22 = df['Revenue 2022'].values
rev_23 = df['Revenue 2023'].values
total = df['Total'].values
width = .3

fig, ax = plt.subplots(figsize=(12,6))


bar1 = plt.bar(x-.3, rev_22, width, color='r', label='2022', zorder=2 ,alpha=.8)
bar2 = plt.bar(x, rev_23, width, color='g', label='2023',zorder=2, alpha=.8)
bar3 = plt.bar(x+.3, total, width, color='b',label='Total', zorder=2, alpha=.8)

plt.xticks(x, df.index, rotation=90)
plt.ylim(0,3000)

plt.subplots_adjust(bottom=0.15)

for bars in [bar1, bar2, bar3]:
    plt.bar_label(bars,color='w', fontweight='bold', fontsize=8, padding=-23, rotation=90)

plt.legend()

plt.title('Revenue\n£ Millions', loc='left', fontweight='bold')
plt.grid()
plt.show()
```
