```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
pd.set_option('display.max_columns', None)
df = pd.read_csv('/Users/cemakay/DataScience/Jupyter/Data/pl_20-21.csv', index_col='Unnamed: 0').convert_dtypes()
df
```

```python
#Exclude some columns
df.columns
drop_cols = [
       'Tackle success %', 'Last man tackles', 'Blocked shots',
       'Interceptions', 'Clearances', 'Headed Clearance',
       'Clearances off line', 'Recoveries', 'Duels won', 'Duels lost',
       'Successful 50/50s', 'Aerial battles won', 'Aerial battles lost',
       'Crosses','Cross accuracy %','Shooting accuracy %','Punches',
       'High Claims', 'Catches', 'Sweeper clearances', 'Throw outs',
       'Goal Kicks']

df = df.drop(columns=drop_cols)
```

```python
#Exclude players with 0 appereances
df = df[df.Appearances != 0]
```

```python
df['Goal Involvements'] = df['Goals'] + df['Assists']
```

```python
df.nlargest(5, 'Goal Involvements')
```

```python
df.nlargest(5, 'Appearances')
```

```python
df = df.drop([137,494])
```

```python
app = df.nlargest(5, 'Goal Involvements')

fig = px.bar(app,
             x=app.Name, 
             y=[app.Appearances, 'Goal Involvements'],
             barmode='group',
             text_auto=True,
             height=600,
             labels={'value': 'Count'},
             color_discrete_sequence=["gray","deepskyblue"],
             title="Goal Involvements (Goals + Assists) - \nSeason: 16/17",
             )

fig.update_xaxes(tickangle=270)
fig.show()

```

```python
ppm = df.nlargest(20, 'Passes per match')

fig = px.bar(ppm,
             x=ppm.Name,
             y=ppm['Passes per match'],
             color=ppm['Passes per match'],
             text_auto=True,
             color_continuous_scale='greens',
             height=600)




fig.update_xaxes(tickangle=270)
fig.show()
```

```python
df['punishment'] = df['Red cards'] + df['Yellow cards']
```

```python
pun = df.nlargest(20, 'punishment')

fig = px.bar(pun,
             x=pun.Name,
             y=['Red cards', 'Yellow cards'],
             color_discrete_sequence=['red', 'yellow'],
             barmode='stack',
             text_auto=True,
             title='Red and Yellow cards',
             height=600)

fig.update_xaxes(tickangle=320)
fig.show()

```
